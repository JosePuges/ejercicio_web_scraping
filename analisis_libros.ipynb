# 📊 Análisis de libros (CSV)
Este notebook contiene 4 consultas sobre el dataset y 2 gráficos.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('libros.csv')
df.head()
```

## Consulta 1: Número total de libros

```python
len(df)
```

## Consulta 2: Precio promedio de los libros

```python
df['precio_num'].mean()
```

## Consulta 3: Conteo de libros por categoría (con gráfica)

```python
conteo = df['rating'].value_counts()
conteo.plot(kind='bar')
plt.title('Libros por rating')
plt.xlabel('rating')
plt.ylabel('Cantidad')
plt.show()
```

## Consulta 4: Distribución de precios (histograma)

```python
df['precio_num'].plot(kind='hist', bins=10, rwidth=0.8)
plt.title('Distribución de precios')
plt.xlabel('Precio_num')
plt.ylabel('Frecuencia')
plt.show()
```
